```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as ny
from sklearn import linear_model, preprocessing
```

```python
data = pd.read_csv("car.data")
print(data.head())
```

```python
# need to convert non-numerical data to numerical data since we need to do computation on data
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"])) 
# get the values of the column of buying and turning them into a list
# transform them into appropriate integer value
# return us a numpy array
print(buying)
```

```python
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))
```

```python
predict = "class"

# create x list and y list
X = list(zip(buying, maint, door, persons, persons, safety)) 
#zip creates a bunch of tuple objects with all the different values corresponding with the lists we give it
y = list(cls)
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
```

```python
print(x_train, y_test)
```

```python
# create classifier
model = KNeighborsClassifier(n_neighbors=3)
# fit the model
model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print(acc)
```

```python
# compare the prediction with the actual data to see how well we are actually doing

predicted = model.predict(x_test)

for x in range(len(predicted)):
    print("Predicted: ", predicted[x],"Data: ", x_test[x],"Actual: ",y_test[x])
    n = model.kneighbors([x_test[x]], 3, True)
    # has to put x_test[x] inside a [] because it doesn't know how to look at data that is not two-dimensional
    print("N: ",n)
```
